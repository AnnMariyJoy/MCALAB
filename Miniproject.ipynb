```python
import pandas as pd
```

```python
df = pd.read_csv("/content/heart_attack_prediction_dataset.csv")
df
```

```python
df.isnull().sum()
```

```python
df['Sex'] = df['Sex'].map({'Male': 1, 'Female': 0})  # Simple binary encoding
```

```python
# Split the 'Blood Pressure' column into 'Systolic' and 'Diastolic'
df[['Systolic', 'Diastolic']] = df['Blood Pressure'].str.split('/', n=1, expand=True)


# Convert the new columns to numeric values
df['Systolic'] = pd.to_numeric(df['Systolic'])
df['Diastolic'] = pd.to_numeric(df['Diastolic'])
```

```python
df = df.drop('Blood Pressure', axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['Age', 'Cholesterol', 'Systolic', 'Diastolic','Cholesterol','Heart Rate','Diabetes','Family History']] = scaler.fit_transform(df[['Age', 'Cholesterol', 'Systolic', 'Diastolic','Cholesterol','Heart Rate','Diabetes','Family History']])
```

```python
df
```

```python
x = df[['Age', 'Cholesterol', 'Sex', 'Systolic','Diastolic','Cholesterol','Heart Rate','Diabetes','Family History']] # If the correct column name is 'BloodPressure'
x

```

```python
y=df['Heart Attack Risk']
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.neighbors import KNeighborsClassifier
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

```

```python
# Standardize the features
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
pca = PCA(n_components=0.95)  # Keep 95% of variance
x_train = pca.fit_transform(x_train)
x_test = pca.transform(x_test)
```

```python
# Hyperparameter tuning: Find the best `k`
best_accuracy = 0
best_k = 1
for k in range(1, 200):
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(x_train, y_train)
    y_pred = knn_model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f'k={k}: Accuracy={accuracy:.2f}')
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_k = k

print(f'Best k: {best_k} with Accuracy: {best_accuracy:.2f}')
```

```python
# Train the model
knn_model = KNeighborsClassifier(n_neighbors=10)
knn_model.fit(x_train, y_train)
```

```python
# Make predictions
y_pred = knn_model.predict(x_test)
```

```python
x_test
```

```python
# Evaluate the model
print(f'Accuracy: {accuracy_score(y_test, y_pred)}')
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred))
print('Classification Report:')
print(classification_report(y_test, y_pred))
```
