```python
import matplotlib.pyplot  as plt
import numpy as np
from sklearn import datasets,linear_model 
from sklearn.metrics import mean_squared_error,r2_score
```

```python
df=datasets.load_diabetes()
df['feature_names']
```

```python
diabetes_X,diabetes_y = datasets.load_diabetes(return_X_y=True)
diabetes_X.shape
```

```python
diabetes_y.shape
```

```python
diabetes_X = diabetes_X[:,np.newaxis,2]
diabetes_X.shape
```

```python
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]

diabetes_y_train = diabetes_y[:-20]
diabetes_y_test = diabetes_y[-20:]
```

```python
regr = linear_model.LinearRegression()

regr.fit(diabetes_X_train,diabetes_y_train)
```

```python
diabetes_y_pred = regr.predict(diabetes_X_test)
```

```python
print("Coefficients: \n",regr.coef_)
print("Mean squared error:%2f" % mean_squared_error(diabetes_y_test,diabetes_y_pred))
print("Coefficien of determination: %2f" % r2_score(diabetes_y_test,diabetes_y_pred))
```

```python
plt.scatter(diabetes_X_test,diabetes_y_test,color="black")
plt.plot(diabetes_X_test,diabetes_y_pred,color="blue",linewidth=3)
plt.xlabel("age")
plt.ylabel("diabetes progression")
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
diabetes_X,diabetes_y=datasets.load_diabetes(return_X_y=True)
diabetes_X.shape
diabetes_X=diabetes_X[:,[0,2]]
diabetes_X.shape
```

```python
diabetes_X_train=diabetes_X[:-20]
diabetes_X_test=diabetes_X[-20:]
diabetes_y_train=diabetes_y[:-20]
diabetes_y_test=diabetes_y[-20:]
```

```python
regr=linear_model.LinearRegression()
regr.fit(diabetes_X_train,diabetes_y_train)
```

```python
diabetes_y_pred=regr.predict(diabetes_X_test)
```

```python
print("Cofficients:\n",regr.coef_)
print("Intercet:\n",regr.intercept_)
print("Mean squared error:%.2f" % mean_squared_error(diabetes_y_test,diabetes_y_pred))
print("Cofficient of determination:%.2f" % r2_score(diabetes_y_test,diabetes_y_pred))
```
