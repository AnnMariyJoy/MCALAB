```python
import numpy as np

def create_matrix(mc):
    print(f"\nARRAY {mc} Elements: ")
    array_1 = map(int, input().split())
    array_1 = np.array(list(array_1))
    
    print(f"\nARRAY {mc}, ROW COLUMN: ")
    row, column = map(int, input().split())
    
    if len(array_1) != (row * column):
        print("\nRow and Column size do not match with total elements! Please retry.")
        return create_matrix(mc)
    
    array_1 = array_1.reshape(row, column)
    print(f"\nARRAY {mc}:")
    print(array_1)
    return array_1

# Create the two matrices
arr1 = create_matrix(1)
arr2 = create_matrix(2)

# Check dimensions and calculate dot product if they match
if arr1.shape[1] == arr2.shape[0]:  # Check if columns of arr1 match rows of arr2
    print("\nDot product:")
    print(np.dot(arr1, arr2))
else:
    print("\nDimensions do not match for dot product!")
```
