```python
import matplotlib.pyplot as plt
import numpy as np

men = [22, 30, 35, 35, 26]
women = [25, 32, 30, 35, 29]
labels = ['Group 1', 'Group 2', 'Group 3', 'Group 4', 'Group 5']
x = np.arange(len(labels)) 
bar_width = 0.4  

fig, ax = plt.subplots()
bars1 = ax.bar(x - bar_width/2, men, width=bar_width, label='Men', color='red')
bars2 = ax.bar(x + bar_width/2, women, width=bar_width, label='Women', color='green')

ax.set_xlabel('Groups')
ax.set_ylabel('Values')
ax.set_title('Comparison of Men and Women')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```
